## 匯入資料至系統內定的User


```python
users.get_values() #list data
```

```python
# Django系統內定的User  有超多欄位 我們只用到 編號 id 姓名username 密碼
from django.contrib.auth.models import User
```

```python
# 一行一行寫入
for line in users.get_values():
    u = User() #產生一個新的物件
    u.id = line[0]
    u.username = line[1]
    u.set_password('mis12345')
    u.save()
```

```python
# 看看有多少筆寫入
User.objects.count()
```

```python
User.objects.all()
```

```python
# 印出每一筆
for row in User.objects.all():
    print(row.id, row.username, row.password)
```

## 匯入Rating資料

```python
from engine.models import Rating
```

```python
# 先看 Band資料表 有幾筆資料?
Rating.objects.count()
```

```python
import datetime
```

```python
from engine.models import Movie
```

```python
# 一行一行寫入
for line in ratings.get_values():
    r = Rating() #產生一個新的物件   
    r.movie = Movie.objects.get( id = line[1]) #連結到另外一個表
    r.user = User.objects.get( id = line[0] ) #連結到另外一個表
    r.rating = line[2]
    r.pub_date = datetime.datetime.now()
    r.save()
```

```python
# 看看有多少筆寫入
Rating.objects.count()
```

```python
# 印出每一筆
for row in Rating.objects.all():
    print(row.user.username, row.movie.name, row.rating, row.pub_date)
```
